 2.Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation.

```python
mean =df.TotalPay.mean()
median = df.TotalPay.median()
mode = df.TotalPay.mode()[0]
max = df.TotalPay.max()
min = df.TotalPay.min()
range = max - min
standard_deviaton = df.TotalPay.std()
print('Mean : ',mean)
print('Median : ',median)
print('Mode : ',mode)
print('Min : ',min)
print('Max : ',max)
print('the range of salaries : ',range)
print('Standard deviation :',standard_deviaton)
```

3.Handle missing data by suitable method with explain why you use it.

```python
#because all rows is nall
df.drop(['Notes','Status'],axis = 1,inplace=True)
```

```python
#because nall is (36163/148654) 24.32% of the rows
df.drop(['Benefits'],axis = 1,inplace=True)
```

```python
#delete the remaining nall rows in data
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

4.Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=10, edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(18, 16))
plt.figure(figsize=(10, 6))
department_counts = df['JobTitle'].value_counts()
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%')
plt.title('Department Distribution')
plt.show()
```

5.Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.

```python
grouped_data_basepay = df.groupby('BasePay')
grouped_data_basepay.agg(['mean','median','min','max','std'])
```

```python
grouped_data_year = df.groupby('Year')
grouped_data_year.agg(['mean','median','min','max','std'])
```

```python
grouped_data_over_other = df.groupby(['OvertimePay','OtherPay'])
grouped_data_over_other.agg(['mean','median','min','max','std'])
```

```python
grouped_data_jobtitle = df.groupby('JobTitle')
grouped_data_jobtitle.agg(['mean','median','min','max','std'])
```

```python
average_basepay = grouped_data_basepay.aggregate('mean')
average_basepay
```

```python
average_year = grouped_data_year.agg(['mean'])
average_year
```

```python
average_over_other = grouped_data_over_other.agg(['mean'])
average_over_other
```

```python
average_jobtitle = grouped_data_jobtitle.agg(['mean'])
average_jobtitle
```
